Importing necessary libraries 

```python
import pandas
import seaborn 
import matplotlib.pyplot as mp
```

Read the Iris dataset from the CSV file


```python
iris_data = pandas.read_csv("./Iris.csv")
```

 1: How many tuples does the Iris dataset have?

```python
number_of_tuples = len(iris_data)
print("Number of tuples:", number_of_tuples)
```

2: How many features does the Iris dataset have? Print feature names.


```python
number_of_features = len(iris_data.columns) - 1  # not including the target column ('Class')
feature_names = iris_data.columns[:-1]
print("Number of features:", number_of_features)
print("Feature names:", feature_names.tolist())
```

3: What's the data type for each feature?

```python
data_types = iris_data.dtypes
print("Data types for each feature:")
print(data_types)
```

4: How many of each Classes do you have?

```python
Classes_counts = iris_data['Class'].value_counts()
print("Number of each Class:")
print(Classes_counts)
```

5: Calculate statistics for petal and sepal measurements.

```python
measurement_columns_names = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width']
measurement_stats = iris_data[measurement_columns_names].describe()
print("statistics for measurements:")
print(measurement_stats)
```

6: Make a scatter plot of sepal_length vs sepal_width.

```python
# using seaborn library instead of panda's because it does the same in fewer lines of code
mp.figure(figsize=(10, 8))
seaborn.scatterplot(x='Sepal_length', y='Sepal_width', data=iris_data, hue='Class')
mp.title("Scatter Plot of Sepal Length vs Sepal Width")
mp.show()
```

7: Create four individual histogram plots.

```python
iris_data[measurement_columns_names].hist(figsize=(14, 10))
mp.suptitle("Histograms of Features", y=1.02)
mp.show()
```

8: Create four individual boxplots.

```python
mp.figure(figsize=(14, 10))
num = 1
for i, column in enumerate(measurement_columns_names):
    mp.subplot(2, 2, i + 1)
    seaborn.boxplot(x='Class', y=column, data=iris_data)
    mp.title(f"{num} Boxplot of {column} by Classes")
    num += 1
mp.tight_layout()
mp.show()
```

9: Make a scatter plot matrix to examine the correlation between each of the measurements.

```python
# Below is a code using Panda library 
measurements = iris_data[measurement_columns_names]
pandas.plotting.scatter_matrix(measurements, figsize=(14, 10), diagonal='hist', marker='o', alpha=0.6)
mp.suptitle("Scatter Plot Matrix by Classes", y=1.02)
mp.tight_layout()
mp.show()

# It can also be done using Seaborn library the code is below
# seaborn.pairplot(iris_data, hue='Class', diag_kind='kde')
# mp.suptitle("Scatter Plot Matrix by Classes", y=1.02)
# mp.show()
```
