Importing necessary libraries 

```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn 
```

Read the Iris dataset from the CSV file


```python
iris_df = pd.read_csv("./iris.csv")
```

 1: How many tuples does the Iris dataset have?

```python
num_tuples = len(iris_df)
print("a) Number of tuples:", num_tuples)
```

2: How many features does the Iris dataset have? Print feature names.


```python
num_features = len(iris_df.columns) - 1  # Excluding the target column ('Class')
feature_names = iris_df.columns[:-1]
print("b) Number of features:", num_features)
print("Feature names:", feature_names.tolist())
```

3: What's the data type for each feature?

```python
data_types = iris_df.dtypes
print("c) Data types for each feature:")
print(data_types)
```

4: How many of each species do you have?

```python
species_counts = iris_df['Class'].value_counts()
print("d) Number of each species:")
print(species_counts)
```

5: Calculate statistics for petal and sepal measurements.

```python
measurement_columns = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width']
measurement_stats = iris_df[measurement_columns].describe()
print("e) Summary statistics for measurements:")
print(measurement_stats)
```

6: Make a scatter plot of sepal_length vs sepal_width.

```python
mp.figure(figsize=(8, 6))
seaborn.scatterplot(x='Sepal_length', y='Sepal_width', data=iris_df, hue='Class')
mp.title("f) Scatter Plot of Sepal Length vs Sepal Width")
mp.show()
```

7: Create four individual histogram plots.

```python
iris_df[measurement_columns].hist(figsize=(12, 8))
mp.suptitle("g) Histograms of Features", y=1.02)
mp.show()
```

8: Create four individual boxplots.

```python
mp.figure(figsize=(12, 8))
num = 1
for i, column in enumerate(measurement_columns):
    mp.subplot(2, 2, i + 1)
    seaborn.boxplot(x='Class', y=column, data=iris_df)
    mp.title(f"{num}) Boxplot of {column} by Class")
    num += 1
mp.tight_layout()
mp.show()
```

9: Make a scatter plot matrix to examine correlations.

```python
seaborn.pairplot(iris_df, hue='Class', diag_kind='kde')
mp.suptitle("i) Scatter Plot Matrix and KDE for Measurements by Class", y=1.02)
mp.show()
```
